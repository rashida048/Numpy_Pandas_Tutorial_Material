```python
import pandas as pd
df = pd.read_csv("StudentsPerformance.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
df1 = df.set_index(['race/ethnicity', 'parental level of education', 'lunch',
       'test preparation course', 'gender'])
df1.head()
```

```python
df1.groupby(level = 0).mean()
```

```python
df1.groupby(level = [0, 4]).mean()
```

```python
df1.groupby(level = ['race/ethnicity', 'gender']).mean()
```

```python
df1.groupby(level = ['race/ethnicity', 'gender'])['math score', 'writing score'].mean()
```

```python
df1.groupby(level = ['race/ethnicity', 'gender'])['math score', 'writing score'].agg(['mean', 'median'])
```

```python
df1.groupby(level = ['race/ethnicity', 'parental level of education', 'gender'])['math score', 'writing score'].agg(
    {
        'math score': ['mean', 'median'],
        'writing score': ['min', 'max', 'std']
    }
)
```
