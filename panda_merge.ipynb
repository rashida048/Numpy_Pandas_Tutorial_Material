```python
import pandas as pd
```

```python
df1 = pd.DataFrame(
{
    "Name": ['Mike', "Dave", "Farah", "Rony", "Laura"],
    "Age": [21, 19, 17, 22, 26],
    "City": ["Houston", "Aukland", "Miami", "Sarasota", "Phoenix"]
})
df1
```

```python
df2 = pd.DataFrame(
{
    "Name": ['Mike', "Dave", "Tommy", "Farah", "Rony"],
    "Score": [78, 98, 82, 86, 88],
    "Weight": [190, 230, 153, 120, 160]
})
df2
```

```python
pd.merge(df1, df2, on = "Name")
```

```python
pd.merge(df1, df2, on = "Name", how = "left")
```

```python
pd.merge(df1, df2, on = "Name", how = "right")
```

```python
pd.merge(df1, df2, on = "Name", how = "inner")
```

```python
pd.merge(df1, df2, on = "Name", how = "outer")
```

```python
df3 = pd.DataFrame(
{
    "First Name": ['Mike', "Dave", "Farah", "Rony"],
    "Last Name": ["Flan", "Smith", "Oniel", "Farnandez"],
    "Age": [21, 19, 17, 22],
    "City": ["Houston", "Aukland", "Miami", "Sarasota"],
})
df3
```

```python
df4 = pd.DataFrame(
{
    "First Name": ['Mike', "Dave", "Dave", "Rony"],
    "Last Name": ["Flan", "Smith", "Smith", "Brooks"],
    "Score": [78, 98, 86, 88],
    "Weight": [190, 230, 120, 160]
})
df4
```

```python
pd.merge(df3, df4, on = ["First Name", "Last Name"])
```

```python
pd.merge(df3, df4, on = ["First Name", "Last Name"], how = "left")
```

```python
pd.merge(df3, df4, on = ["First Name", "Last Name"], how = "right")
```

```python
pd.merge(df3, df4, on = ["First Name", "Last Name"], how = "inner")
```

```python
pd.merge(df3, df4, on = ["First Name", "Last Name"], how = "outer")
```
