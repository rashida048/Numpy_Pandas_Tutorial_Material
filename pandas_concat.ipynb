Concatenation of Pandas Series and DataFrames

```python
import pandas as pd
```

```python
c1 = pd.Series([10, 12, 14, 12, 11], name = "col1")
c2 = pd.Series([21, 13, 10, 6, 7])
c3 = pd.Series([34, 32, 31, 10, 15], name = "col3")
c4 = pd.Series([21, 25, 29, 10, 16])
```

```python
pd.concat([c1, c2, c3, c4])
```

```python
pd.concat([c1, c2, c3, c4], axis = 1)
```

```python
pd.concat([c1, c2, c3, c4], axis = 1, keys = ['A', 'B', 'C', 'D'])
```

```python
df1 = pd.DataFrame({
    "Name": ["Ria", "Lia", "John", "Kevin", "Lily"],
    "SN": [12, 8, 30, 21, 19],
    "Position": ["Fifth", "Second", "First", "Tenth", "Twelveth"],
    "Group" : ["B", "D", "A", "C", "E"]
}, index = [0, 1, 2, 3, 4]
)
```

```python
df1
```

```python
df2 = pd.DataFrame({
    "Name": ["Laura", "Jeff", "Rony", "Lian"],
    "SN": [34, 20, 3, 14],
    "Position": ["Third", "Twenty First", "Fifteenth", "Seventh"],
    "Group" : ["C", "F", "A", "E"]
}, index = [5, 6, 7, 8]
)
```

```python
df2
```

```python
df3 = pd.DataFrame({
    "Name": ["Mike", "Molly", "Jenny", "Rob", "Clara"],
    "SN": [22, 31, 10, 17, 27],
    "Position": ["Sixth", "Seventeenth", "Twenty Second", "Forth", "Ninth"],
    "Group" : ["D", "F", "A", "c", "E"]
}, index = [9, 10, 11, 12, 13]
)
```

```python
df3
```

```python
pd.concat([df1, df2, df3])
```

```python
con = pd.concat([df1, df2, df3], keys = ["A", "B", "C"])
con
```

```python
con.loc["A"]
```

```python
con.loc[["A", "C"]]
```

```python
d = {"A": df1, "B": df2, "C": df3}
```

```python
d
```

```python
pd.concat(d)
```

```python
df_con = pd.concat(d, keys = ["B", "A", "C"])
df_con
```

```python
df_con.loc[["A", "C"]]
```

```python
df_con.index.levels
```

```python
pd.concat(d, keys = ["B", "A", "C"], names = ["Section", "N"])
```
